```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10,6)
```

```python
with open('/content/CA_category_id.json') as f:
    categories = json.load(f)

cat_data = []
for item in categories['items']:
    cat_id = item['id']
    cat_name = item['snippet']['title']
    cat_data.append({'category_id': int(cat_id), 'category_name': cat_name})

cat_df = pd.DataFrame(cat_data)
cat_df.head()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('/content/CAvideos.csv')
df.head()
```

```python
df = df.merge(cat_df, how='left', on='category_id')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
df.dropna(subset=['title', 'channel_title'], inplace=True)
df['comment_count'] = df['comment_count'].fillna(0)
df.drop_duplicates(inplace=True)
df = df[df['views'] > 0]
df = df[df['likes'] <= df['views']]
```

```python
sns.histplot(df['views'], bins=50, kde=True)
plt.title('Views Distribution')
plt.xlabel('Number of Views')
plt.ylabel('Number of Videos')
plt.show()
```

```python
sns.countplot(y='category_name', data=df,
order=df['category_name'].value_counts().index)
plt.title('Number of videos by category')
plt.xlabel('Number of videos')
plt.ylabel('Category name')
plt.show()
```

```python
sns.scatterplot(x='views', y='likes', data=df)
plt.title('Relationship between views and likes')
plt.xlabel('Number of views')
plt.ylabel('Number of likes')
plt.show()
```

```python
df.groupby('category_name')['comment_count'].mean().sort_values(ascending=False).plot(kind='barh')
plt.title('Average number of comments by category')
plt.xlabel('Average comments')
plt.ylabel('Category name')
plt.show()
```

```python
df.to_csv('/content/CAvideos_cleaned.csv', index=False)
```
